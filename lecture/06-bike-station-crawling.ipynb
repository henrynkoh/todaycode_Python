```python
def find_tds(tr):
    tds = tr.select('td')
    info = []
    for i, td in enumerate(tds) :
        if i == 0:
            latlong = td.find('a')['param-data']
            latlong = latlong.split(',')
            lat = latlong[0]
            long = latlong[1]
            station = td.get_text(strip=True)
            station = station.split('.')
            station_id = station[0]
            station_name = station[-1]
            
            if len(station) == 2:
                station_id = station[0]
                station_name = station[-1].strip()
            else:
                station_id = np.nan
                station_name = station[-1].strip()

            info.append(station_id)
            info.append(station_name)
            info.append(lat)
            info.append(long)
        else:
            info.append(td.get_text(strip=True))
            
    return info

find_tds(trs[3])
```

```python
# 전체 대여소 정보를 담아줄 비어있는 리스트를 만듭니다.
all_stations = []

def crawling_rent_station(pnum):
    
    base_url = 'https://www.bikeseoul.com/app/station/moveStationSearchView.do?currentPageNo='
    response = requests.get(base_url+str(pnum))

    html = bs(response.text, 'html.parser')
    
    table = html.select('#container > table')[0]
    trs = table.select('tr')
    
    if len(trs) > 1 :
        stations = station_info(trs)
        all_stations.extend(stations)
        pnum = pnum + 1
        # 재귀호출
        s = random.uniform(1, 5)
        time.sleep(s)
        return crawling_rent_station(pnum)
    else:
        return all_stations
    
crawling_rent_station(308)
```

## 각 대여소의 정보를 받아오기
1. tbody 값이 있는지 확인합니다.
1. tr을 반복문으로 순회합니다.
1. '.' 이 대여소명에 들어가는지 확인하고 이 문자로 대여소 번호와 대여소명을 가져옵니다. 
1. get_text()를 통해 문자만 가져올 수 있습니다.
1. 없다면 대여소번호를 np.nan 으로 결측치를 넣어주고 대여소는 텍스트 데이터를 넣어줍니다.
1. 거치대수를 가져옵니다.
1. 운영여부를 가져옵니다.
1. 주소를 가져옵니다.
1. 위도, 경도를 가져옵니다 .find('a')['param-data'] 로 찾습니다.
1. 위도, 경도를 "," 문자로 분리합니다.
1. 대여소 리스트에 info 리스트를 담습니다.
1. rent_stations 에 담고 다 만들었다면 반환합니다.

```python
# ['대여소번호', '대여소', '거치대수', '상태', '주소', '위도', '경도'] 를 반환합니다.
# trs를 받아 2차원 리스트로 반환해 주는 함수를 만듭니다.
def station_info(trs):
    
    rent_stations = []
    for i, tr in enumerate(trs):
        if i > 0 :
            row = find_tds(tr)
            rent_stations.append(row)

    return rent_stations
```
